```python
import re
from num2words import num2words

# Define a mapping of symbols to words
symbol_to_word = {
    '%': ' percent',
    '$': 'dollar ',
    '&': 'and',
    '@': ' at '
}

# Function to convert text
def convert_text(text):
    # Replace symbols with words
    for symbol, word in symbol_to_word.items():
        text = text.replace(symbol, word)

    # Replace numbers with words
    text = re.sub(r'\b\d+\b', lambda x: num2words(int(x.group())), text)

    # Remove brackets
    text = re.sub(r'[\[\]{}()]', '', text)

    return text
```
